```python
import pandas as pd
dataset=pd.read_csv("kidney_disease.csv")
dataset
```

```python
def quanQual(dataset):
    quan=[]
    qual=[]
    for columnName in dataset.columns:
        if(dataset[columnName].dtype=='0'):
            qual.append(columnName)
        else:
            quan.append(columnName)
    return quan,qual
        
```

```python
quan,qual=quanQual(dataset)
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imp.fit(dataset[quan])
df = imp.transform(dataset[quan])
#df=pd.DataFrame(dfcolumns=quan)
```

```python
# Listing the columns with NaN values
missing_values_count = dataset.isnull().sum()
#tolist() method is used to convert array objects to lists
#missing_values_count > 0: creates a boolean series where 'True' indicates a column with at least one missing value
#missing_values_count[...]:  uses boolean indexing to filter the missing_values_count series, keeping only the rows where the count is greater than zero.
#.index: retrieves the index  of the filtered series.
columns_with_nan = missing_values_count[missing_values_count > 0].index.tolist()
print("Columns with NaN values:", columns_with_nan)
```

```python
#dataset.select_dtypes(include=np.number): selects a subset of columns from the DataFrame based on their data types. 
#include=np.number specifies that we want to select all columns that have a numeric data type (e.g., int64, float64).
numeric_cols = dataset.select_dtypes(include=np.number).columns.tolist()
#dataset.select_dtypes(include='object'): selects all columns with a categorical or string-like data type.
#.columns.tolist(): creates a list of the names of all the categorical columns, stored in categorical_cols.
categorical_cols = dataset.select_dtypes(include='object').columns.tolist()
```

```python
# Impute numeric columns with 'mean'
from sklearn.impute import SimpleImputer
imputer_numeric = SimpleImputer(strategy='mean')
#fit(): calculates the mean value for each of the numeric columns
#transform(): the calculated means are filled in the missing NaN values in each respective column
dataset[numeric_cols] = imputer_numeric.fit_transform(dataset[numeric_cols])

# Impute categorical columns with 'most frequent' value
#strategy='most_frequent' is suitable for categorical data, as it replaces missing values with the most common value (mode) in that column
imputer_categorical = SimpleImputer(strategy='most_frequent')
#imputer_categorical.fit_transform(...): this calculates the most frequent value for each categorical column and then uses those values to fill the missing NaNs
#The imputed data is then assigned back to the categorical columns in the dataset
dataset[categorical_cols] = imputer_categorical.fit_transform(dataset[categorical_cols])
```
